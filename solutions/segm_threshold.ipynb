**Task 1.2.a.** Use `imread` to load an image:

```python
img = imread('../data/NIH3T3/im/dna-0.png')
```

**Task 1.2.b.** Show the image:

```python
figure()
imshow(img, 'gray')
colorbar()
```

**Task 1.2.d.** Binarize the image by applying an intensity threshold:

```python
seg = (img >= 0.1)
imshow(seg, 'gray')
```

**Task 1.2.e.** Load the ground truth:

```python
gt = imread('../data/NIH3T3/gt/0.png').astype(bool)
```

Compute and print the Dice coefficient:

```python
dice = 2 * (gt * seg).sum() / (gt.sum() + seg.sum())
print(dice)
```

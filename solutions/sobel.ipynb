**Task 2.2.** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    img_filtered = numpy.zeros(shape=img.shape)
    
    gx = 1/8 * array(([-1, 0, 1],
                      [-2, 0, 2],
                      [-1, 0, 1]))
    
    for p in numpy.ndindex(img.shape):
        if 0 < p[0] < img.shape[0] - 1 and 0 < p[1] < img.shape[1] - 1:
            value = sum(img[p[0] - 1 : p[0] + 2, p[1] - 1 : p[1] + 2] * gx)
            img_filtered[p[0], p[1]] = value
            
    return img_filtered
    
def sobel_v(img):
    img_filtered = numpy.zeros(shape=img.shape)

    gy = 1/8 * numpy.array(([-1, -2, -1],
                      [ 0,  0,  0],
                      [ 1,  2,  1]))
    
    for p in numpy.ndindex(img.shape):
        if 0 < p[0] < img.shape[0] - 1 and 0 < p[1] < img.shape[1] - 1:
            value = sum(img[p[0] - 1 : p[0] + 2, p[1] - 1 : p[1] + 2] * gy)
            img_filtered[p[0], p[1]] = value
            
    return img_filtered
```

**Task 2.3.** Test your implementations from above:

```python
figure()
imshow(sobel_h(img))
title('sobel_h')
colorbar()

figure()
imshow(sobel_v(img))
title('sobel_v')
colorbar()
```

**Task 2.4.** Compute the image gradient magnitude:

```python
dgx = sobel_h(img)
dgy = sobel_v(img)
gradmag = sqrt(dgx ** 2 + dgy ** 2)

figure()
imshow(gradmag)
colorbar()
```
